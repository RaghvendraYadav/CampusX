## Filtering a DataFrame

```python
ipl.head(2)
```

```python
mask=ipl['MatchNumber']=='Final'
```

```python
new_df=ipl[mask]
```

```python
new_df[['WinningTeam','Season']]
```

```python
#every years winner name
ipl[ipl['MatchNumber']=='Final'][['WinningTeam','Season']]
```

```python
#how many super over have occured
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
#how many matches has csk won in kolkata
ipl[(ipl['WinningTeam']=='Chennai Super Kings') & (ipl['City']=='Kolkata')].shape[0]
```

```python
#toss winner is match winner in percentage
(ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
#movies with higher rating than 8 and votes>100000
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
#Action movies rating greater >7.5
filt1=movies['genres'].str.split('|').apply(lambda x:'Action' in x)
filt2=movies['imdb_rating']>7.5
movies[filt1 & filt2]
```
