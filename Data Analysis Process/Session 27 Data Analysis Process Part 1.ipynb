```python
def  rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
#converters
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename,
                                                   'team2':rename,
                                                   'toss_winner':rename,
                                                   'toss_decision':rename,
                                                   'winner':rename})
```
