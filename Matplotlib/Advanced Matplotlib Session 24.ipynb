```python
fig,ax=plt.subplots(figsize=(15,10))
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
#adding title in subplots using set_title
ax.set_title('Performance of batsman ')
#xlabel
ax.set_xlabel('Average')
#ylabel
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
#two subplot
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,10))
ax[0].scatter(batters['avg'],batters['runs'])
ax[0].set_title('Average Vs Runs')

ax[0].set_ylabel('Runs')
ax[1].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].set_title('Average Vs Strike Rate')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')
```

```python
#four subplot
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))
ax[0,0].scatter(batters['avg'],batters['runs'])
ax[0,0].set_title('Average Vs Runs')
ax[0,0].set_ylabel('Runs')
ax[0,1].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].set_title('Average Vs Strike Rate')
ax[0,1].set_xlabel('Average')
ax[0,1].set_ylabel('Strike Rate')
ax[1,0].hist(batters['runs'])
ax[1,1].hist(batters['avg'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```

# 3D Surface Plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z=xx**2 + yy**2
z
```

```python
z.shape
```

```python
fig=plt.figure(figsize=(15,10))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+ np.cos(xx)
```

```python
fig=plt.figure(figsize=(15,10))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## contour plot   if we want to show 3d into 2d we use contour plot

```python
fig=plt.figure(figsize=(15,10))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#contourf - fill 

fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+ np.cos(xx)
fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
delivery['ballnumber'].unique()
```

```python
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
temp_df
```

```python
#pivot_table
grid=temp_df.pivot_table(index='overs',values='batsman_run',columns='ballnumber',aggfunc='count')
```

```python
plt.figure(figsize=(15,10))
plt.imshow(grid)
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.colorbar()
```
