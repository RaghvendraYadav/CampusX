```python
#label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
```

```python
#hex code for colors 
plt.plot(batsman['index'],batsman['V Kohli'],color='#33A5FF')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF7133')
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
```

```python
#linestyle  --dashed,dotted,dashdot,solid
#linewidth=3
plt.plot(batsman['index'],batsman['V Kohli'],color='#33A5FF',linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF7133',linestyle='dashdot',linewidth=2)
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
```

```python
#markers-- D(Diamond),P(Plus),.,<,>,o
#markersize=10
plt.plot(batsman['index'],batsman['V Kohli'],color='#33A5FF',linestyle='dashdot',linewidth=3,marker='D',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF7133',linestyle='dashdot',linewidth=2,marker='o',markersize=7)
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs') 
```

```python
#label + plt.legend(loc=),loc=upper right,
plt.plot(batsman['index'],batsman['V Kohli'],color='#33A5FF',linestyle='dashdot',linewidth=3,marker='D',markersize=8,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF7133',linestyle='dashdot',linewidth=2,marker='o',markersize=7,label='rohit')
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
plt.legend(loc='upper right') 
```

```python
#limiting axes -- due to outlier grid
price=[48000,50000,52000,40000,49000,3200000]
years=[2015,2016,2017,2018,2019,2020]
plt.plot(years,price)
plt.ylim(1,70000)
plt.xlim(2017,2019)
plt.grid()
```

```python
#show
plt.plot(batsman['index'],batsman['V Kohli'],color='#33A5FF',linestyle='dashdot',linewidth=3,marker='D',markersize=8,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF7133',linestyle='dashdot',linewidth=2,marker='o',markersize=7,label='rohit')
plt.title('Rohit Sharma Versus Virat Kohli  Career Comparison')
plt.xlabel('Seasons')
plt.ylabel('Total Runs')
plt.legend(loc='upper right') 
plt.grid()
plt.show()
```

```python
df=pd.read_csv('/content/sample_data/batsman_season_record.csv')
```

```python
#multiple Bar Graph  using width


plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.19,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.19,color='green')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.19,color='blue')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# a problem  x axis name 
#xticks
children=[15,12,21,10,25]
color=['Red Red Shree Ram Jai Ram Jai Jai Ram ','Blue Mangal Bhavan A mangal Hari','Green Jai Jai Hanumna gosai ','Yellow Dil m mere  h dard de disco','Pink humko Pine h pine h pine h ']
```

```python
plt.bar(color,children)
plt.xticks(rotation='vertical')
```

```python
plt.bar(color,children)
plt.xticks(rotation=70)
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')

plt.legend()
plt.show()
```
